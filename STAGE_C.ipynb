```python
from sklearn.preprocessing import OrdinalEncoder
enc = OrdinalEncoder()
ordinal_categoricals = ["stabf"]
```

```python
df[ordinal_categoricals]=enc.fit_transform(df[ordinal_categoricals])  #encoding the target varibale into numerical
```

```python
df.head()
#From the encoding, 1.0 is interpreted as unstable while 0.0 is interpreted as stable
```

```python
#seperating the data into target(y) and response variable (X)
X=df.drop(columns=['stabf'])
y= df['stabf']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)  #Splinting the data into training, and testing with test size of 20 percent
```

```python
from sklearn.metrics import accuracy_score, recall_score,precision_score, f1_score
```

**USING XGBOOST**

```python
import xgboost as xgb
```

```python
xgb_classifier = xgb.XGBClassifier()
xgb_classifier.fit(X_train, y_train)
```

```python
y_pred = xgb_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy score is :", round(accuracy, 4))
```


**Using Random Forest Classfier**



```python
from sklearn.ensemble import RandomForestClassifier
RC= RandomForestClassifier()
```

```python
RC.fit(X_train,y_train)
```

```python
yrc_pred= RC.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, yrc_pred)
print("Accuracy score is :", round(accuracy, 4))
```
